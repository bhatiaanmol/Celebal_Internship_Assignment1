 Handling Missing Values

```python
df["discount"] = df["discount"].fillna(0)

df["what_customers_said"] = df["what_customers_said"].fillna("No Review")

df["seller_name"] = df["seller_name"].fillna("Unknown Seller")

df["videos"] = df["videos"].fillna("No Video")

df["seller_information"] = df["seller_information"].fillna("Not Available")

df["variations"] = df["variations"].fillna("No Variations")
```

```python
df.isnull().sum()
```

Duplicate Record Check

```python
print("Duplicates: ",df.duplicated().sum()) 
```

Data Cleaning
- final_price conversion
- datatype correction

```python
df["discount_amount"] = (df["initial_price"] - df["final_price"])
```

```python
print(df["initial_price"].dtype)
print(df["final_price"].dtype)
```

```python
df[["initial_price", "final_price"]].head()
```

```python
df["final_price"] = pd.to_numeric(
    df["final_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False),
    errors="coerce"
)
```

```python
df["discount_amount"] = df["initial_price"] - df["final_price"]
```

```python
df[["initial_price", "final_price", "discount_amount"]].head()
```
